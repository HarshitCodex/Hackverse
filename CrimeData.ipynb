```python
import random
```

```python
def g(j):
    x=random.randint(0,24)
    if(x>=21 and x<24):
        return "Night"
    elif (x>=0 and x<5):
        return "Night"
    elif (x>=5 and x<12):
        return "Morning"
    elif (x>=12 and x<17):
        return "Afternoon"
    elif (x>=17 and x<21):
        return "Evening"
    else :
        return "Morning"
    
```

```python
df_testing['Phase']=df_testing['Month'].apply(g)
```

```python
df_testing.head(20)
```

```python
def t(x):
    if(x[7]=='Night'):
        return x[6]+12
    elif(x[7]=='Morning'):
        return x[6]-3
    elif(x[7]=='Afternoon'):
        return x[6]-12
    elif(x[7]=='Morning'):
        return x[6]-8
    else:
        return x[6]+2
    
```

```python
df_testing['Factor']=df_testing.apply(t,axis=1)
```

```python
df_testing.head(10)
```

```python
def com(x):
    return x[1],x[2],x[7]
```

```python
df_testing['Point']=df_testing.apply(com,axis=1)
```
