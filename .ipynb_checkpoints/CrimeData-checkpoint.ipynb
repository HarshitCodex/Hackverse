```python
import random
```

```python
def f(x):
    if(x.lower()=='Violence and sexual offences'):
        return 90
    elif(x =='Anti-social behaviour'):
        return 20
    elif(x =='Other theft'):
        return 30
    elif(x=='Criminal damage and arson'):
        return 50
    elif(x=='Vehicle crime'):
        return 90
    elif(x=='Burglary'):
        return 60
    elif(x=='Public order'):
        return 60
    elif(x=='Shoplifting'):
        return 30
    elif(x=='Drugs'):
        return 40
    elif(x=='Other crime'):
        return 50
    elif(x=='Robbery'):
        return 60
    elif(x=='Bicycle theft'):
        return 60
    elif(x=='Possession of weapons'):
        return 80
    elif(x=='Theft from the person'):
        return 70
    else:
        return random.randrange(20,50,10)
```

```python
df_raw['Factor']=df_raw['Crime'].apply(f)
```
